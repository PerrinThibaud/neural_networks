# Import

```python
from viznet import NodeBrush, EdgeBrush, DynamicShow
import matplotlib.pyplot as plt
```

## Perceptron (p)

```python
plt.axis("equal")
plt.axis("off")
# with DynamicShow(filename='./test.png', dpi=300, fps=1) as d:
pipe = NodeBrush('qc.end', size='normal')
brushInv = NodeBrush('invisible', size='normal')
brushIn = NodeBrush('nn.input', size='normal')
brushOut = NodeBrush('nn.output', size='normal')
node1 = brushIn >> (1,1)  # paint a node at (x=1, y=0)
node2 = brushIn >> (1,0)
node3 = brushOut >> (2,.5)
pipe >> (2,.5)
node4 = brushInv >> (3,.5)

edge = EdgeBrush('-', lw=2)

edge >> (node1, node3)  # connect two nodes
edge >> (node2, node3)  # connect two nodes
edge >> (node3, node4)  # connect two nodes

node1.text('$x_1$', 'center', fontsize=18)  # add text to node1
node2.text('$x_2$', 'center', fontsize=18)
node3.text('$z$  $a$', 'center', fontsize=18)
node4.text('$y$', 'center', fontsize=18)
plt.savefig('./Perceptron.png', dpi=300, transparent=True)
plt.tight_layout()
```

## Feed Forward (FF)

```python
plt.axis("equal")
plt.axis("off")
# with DynamicShow(filename='./test.png', dpi=300, fps=1) as d:
pipe = NodeBrush('qc.end', size='large')
mpo = NodeBrush('tn.mpo', size='normal', color='#fffaaa')
brushInv = NodeBrush('invisible', size='large')
brushIn = NodeBrush('nn.input', size='large')
brushHide = NodeBrush('nn.hidden', size='large')
brushOut = NodeBrush('nn.output', size='large')

mpo.size = (1, 1.5)
node0 = mpo >> (5, 1.5)
node1 = brushIn >> (1,2)  # paint a node at (x=1, y=0)
node2 = brushIn >> (1,1)
node3 = brushOut >> (5,2)
pipe >> (5,2)
node4 = brushOut >> (5,1)
pipe >> (5,1)
text = brushInv >> (0, 4.5)
text2 = brushInv >> (0, 4)
b1 = brushInv >> (3, 3)
b2 = brushInv >> (3, 0)

edge = EdgeBrush('->', lw=2)
w11 = edge >> (node1, node3)  # connect two nodes
w21 = edge >> (node1, node4)  # connect two nodes
w12 = edge >> (node2, node3)
w22 = edge >> (node2, node4)
edge >> (b1, node3)
edge >> (b2, node4)

# node text
node1.text('$x_1$', 'center', fontsize=18)  # add text to node1
node2.text('$x_2$', 'center', fontsize=18)
node3.text('$z_1$  $a_1$', 'center', fontsize=18)
node4.text('$z_2$  $a_2$', 'center', fontsize=18)
b1.text('$b_1$', 'center', fontsize=18)
b2.text('$b_2$', 'center', fontsize=18)
text.text('$w_{ij}$: paramtre associ au neurone $i$ \n et provenant de l\'entre $j$', 'right', fontsize=18)
text2.text('$b_{i}$: biais associ au neurone $i$', 'right', fontsize=18)
node0.text('Couche[1]', 'top', fontsize=18)

# edge text
w11.text('$w_{11}$', 'top', fontsize=18)
w21.text('$w_{21}$', 'top', fontsize=18)
w12.text('$w_{12}$', 'bottom', fontsize=18)
w22.text('$w_{22}$', 'bottom', fontsize=18)
plt.rcParams["figure.figsize"] = (8,8)
# plt.savefig('./FeedForward.png', dpi=300, transparent=True)
plt.tight_layout()
```

```python
plt.axis("equal")
plt.axis("off")
# with DynamicShow(filename='./test.png', dpi=300, fps=1) as d:
pipe = NodeBrush('qc.end', size='large')
mpo = NodeBrush('tn.mpo', size='normal', color='#fffaaa')
brushInv = NodeBrush('invisible', size='large')
brushIn = NodeBrush('nn.input', size='large')
brushHide = NodeBrush('nn.hidden', size='large')
brushOut = NodeBrush('nn.output', size='large')

mpo.size = (.75, 1.5)
mpo1 = mpo >> (5, 1.5)
mpo2 = mpo >> (9, 1.5)
node1 = brushIn >> (1,2)  # paint a node at (x=1, y=0)
node2 = brushIn >> (1,1)
node3 = brushHide >> (5,2)
pipe >> (5,2)
node4 = brushHide >> (5,1)
pipe >> (5,1)
node5 = brushOut >> (9,1.5)
pipe >> (9,1.5)
b1 = brushInv >> (3, 3)
b2 = brushInv >> (3, 0)
b12 = brushInv >> (8, 3)

edge = EdgeBrush('->', lw=2)
w11 = edge >> (node1, node3)  # connect two nodes
w21 = edge >> (node1, node4)  # connect two nodes
w12 = edge >> (node2, node3)
w22 = edge >> (node2, node4)
w112 = edge >> (node3, node5)
w122 = edge >> (node4, node5)
edge >> (b1, node3)
edge >> (b2, node4)
edge >> (b12, node5)

# node text
node1.text('$x_1$', 'center', fontsize=18)  # add text to node1
node2.text('$x_2$', 'center', fontsize=18)
node3.text('$z^{[1]}_1$  $a^{[1]}_1$', 'center', fontsize=18)
node4.text('$z^{[1]}_2$  $a^{[1]}_2$', 'center', fontsize=18)
node5.text('$z^{[2]}_1$  $a^{[2]}_1$', 'center', fontsize=18)

b1.text('$b^{[1]}_1$', 'center', fontsize=18)
b2.text('$b^{[1]}_2$', 'center', fontsize=18)
b12.text('$b^{[2]}_1$', 'center', fontsize=18)

mpo1.text('Couche[1]', 'top', fontsize=18)
mpo2.text('Couche[2]', 'top', fontsize=18)

# edge text
w11.text('$w^{[1]}_{11}$', 'top', fontsize=18)
w21.text('$w^{[1]}_{21}$', 'top', fontsize=18)
w12.text('$w^{[1]}_{12}$', 'bottom', fontsize=18)
w22.text('$w^{[1]}_{22}$', 'bottom', fontsize=18)
w112.text('$w^{[2]}_{11}$', 'top', fontsize=18)
w122.text('$w^{[2]}_{12}$', 'bottom', fontsize=18)
plt.rcParams["figure.figsize"] = (20,10)
# plt.savefig('./FeedForward2C.png', dpi=300, transparent=True)
plt.tight_layout()
```

```python
plt.axis("equal")
plt.axis("off")
# with DynamicShow(filename='./test.png', dpi=300, fps=1) as d:
pipe = NodeBrush('qc.end', size='normal')
mpo = NodeBrush('tn.mpo', size='normal', color='#fffaaa')
brushInv = NodeBrush('invisible', size='normal')
brushIn = NodeBrush('nn.input', size='normal')
brushHide = NodeBrush('nn.hidden', size='normal')
brushOut = NodeBrush('nn.output', size='normal')

layer1 = []
layer1.append(brushIn >> (1,3.5))
layer1.append(brushIn >> (1,2.5))
layer1.append(brushIn >> (1,1.5))

layer2 = []
layer2.append(brushHide >> (2,4))
layer2.append(brushHide >> (2,3))
layer2.append(brushHide >> (2,2))
layer2.append(brushHide >> (2,1))

layer3 = []
layer3.append(brushHide >> (3,4))
layer3.append(brushHide >> (3,3))
layer3.append(brushHide >> (3,2))
layer3.append(brushHide >> (3,1))

layer4 = []
layer4.append(brushOut >> (4,2))
layer4.append(brushOut >> (4,3))

layers = [layer1, layer2, layer3, layer4]

edge = EdgeBrush('-', lw=2)

for l in range(len(layers)-1):
    for node in layers[l]:
        for linked_node in layers[l + 1]:
            edge >> (node, linked_node)

plt.savefig('./DeepFeedForward.png', dpi=300, transparent=True)
plt.tight_layout()
```
