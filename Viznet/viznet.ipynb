# Import

```python
from viznet import NodeBrush, EdgeBrush, DynamicShow
```

## Perceptron (p)

```python
import matplotlib.pyplot as plt

plt.axis("equal")
plt.axis("off")
# with DynamicShow(filename='./test.png', dpi=300, fps=1) as d:
brushIn = NodeBrush('nn.input', size='normal')
brushOut = NodeBrush('nn.output', size='normal')
node1 = brushIn >> (1,0)  # paint a node at (x=1, y=0)
node2 = brushIn >> (1,2)
node3 = brushOut >> (2,1)
edge = EdgeBrush('-', lw=2)
edge >> (node1, node3)  # connect two nodes
edge >> (node2, node3)  # connect two nodes
node1.text('x1', 'center', fontsize=18)  # add text to node1
node2.text('x2', 'center', fontsize=18)
node3.text('y', 'center', fontsize=18)
plt.savefig('./Perceptron.png', dpi=300, transparent=True)
plt.tight_layout()
```
